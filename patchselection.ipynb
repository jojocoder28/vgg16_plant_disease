```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("./bgblur_dataset/Tomato___Bacterial_spot/0c09c121-e945-4b7e-acbf-dff4e0d01acb___GCREC_Bact.Sp 3379.JPG")
```

```python
gray1 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#keypoints
sift = cv2.SIFT_create()
keypoints_1, descriptors_1 = sift.detectAndCompute(image,None)

img_1 = cv2.drawKeypoints(gray1,keypoints_1,image)
plt.imshow(img_1)
```
